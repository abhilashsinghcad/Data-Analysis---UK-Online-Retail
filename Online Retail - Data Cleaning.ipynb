# Data Cleaning

The dataset is really large. Therefore, we need to clean the data carefully prior running any analysis. Such large datasets tend to skew the results if not handled properly. Initially, we will stick to the basics and attempt to dig deeper based on research and our understanding.

```python
# Concat - Combining the two sheets in the files for different years (2009-2010) and (2010-2011)

df = pd.concat([retail,retail_2])

df.drop_duplicates(inplace = True)  # Duplicate Records

df.reset_index(inplace = True)
df.drop('index', axis = 1, inplace = True)
df
```

```python
df.info()
```

```python
# Lets not remove the missing records for now, instead replace it with some other value.

df['Customer ID']=df['Customer ID'].fillna(-1)
df['Customer ID']=df['Customer ID'].astype('int')
```

```python
# Missing Values
df.isnull().sum()
```

```python
# Cancelled Orders - Does not count towards revenue stream.

cancel_df = df[df['Invoice'].str.contains('C', na = False)]
cancel_df.reset_index(drop=True)
```

```python
# Transactions Income - Counts Towards Revenue.

trans_df = df[~df['Invoice'].str.contains('C', na = False)]

trans_df.loc[:,'Revenue'] = np.nan

trans_df.loc[:,'Revenue']= trans_df.loc[:,'Quantity']*trans_df.loc[:,'Price']

trans_df.reset_index(drop=True)
```

```python
# Out of the transactions made, there are some related to Bad Debt Adjustments.
# Additionally, the Customer ID's were missing for the same transactions (We replaced NaNs with -1).

bad_debt = trans_df[trans_df.Invoice.str.contains('A', na = False)]
bad_debt
```

### What Is a Bad Debt Expense?

"A bad debt expense is recognized when a receivable is no longer collectible because a customer is unable to fulfill their obligation to pay an outstanding debt due to bankruptcy or other financial problems. Companies that extend credit to their customers report bad debts as an allowance for doubtful accounts on the balance sheet, which is also known as a provision for credit losses."

```python
trans_df[trans_df['Price']<0]
```

```python
# Filtering the Adjusted bad debt data and transactions where price equal to 0.

trans_df = trans_df[~trans_df.Invoice.str.contains('A', na = False)]
trans_df = trans_df[trans_df['Price']!=0]
trans_df.reset_index(drop=True)
```
